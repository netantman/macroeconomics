## What is money supply?

The money supply is all the currency and other liquid instruments in a country's economy on the date measured. The money supply roughly includes both cash and deposits that can be used almost as easily as cash. M0, M1 and M2 are the most widely track measure of money supply

- **M0** refers to the most liquid form of money: cash. That includes central bank notes and coins.

## How is money 'made'?

The three players in the money supply process are the central bank, banks (depository institutions), and depositors.

Four items in the central banks balance sheet are essential to our understanding of the money supply process: the two liability items, currency in circulation and reserves, which together make up the **monetary base** (sometimes denoted as **MB**), and the two asset items, securities and loans to financial institutions. More on central banks balance sheet later.

The central bank controls the monetary base through open market operations and extensions of loans to financial institutions (discount loans), both creating reserves for financial institutions (banks or non-banks, such as security dealers). Financial institutions are then free to use these reserves to purchase securities or extend loans, which will end up in depositors' hands. They can either deposit them back to banks (and thus come back as reserves) or becomes currency in circulation. Central banks thus have absolute control over MB, but not necessarily on reserves. Note that there are short-term factors that affect monetary base, such as funds for foreign exchange (in the case of China), floats (i.e. checks that are in the process of being cleared by the central banks) and treasury deposits (inflow or outflow), but central banks can oversee these factors and conteract accordingly. More on the different ways how central banks conduct monetary policy by controling monetary base later.

Due to the fractal reserve system, MB is 'multiplied' when banks extend loans or financial institution purchase securities, i.e. in-balance-sheet activities (in particular, in China, when banks purchase SPVs, it also creates M2). ## Is Money Supply, or M2, 'liquidity'?

The two are related, but not entirely the same.

M2, as described above, is a concept of money supply, while liquidity, especially in the perspective of security transactions, represents how much fund can be and is willing to be allocated for the transaction. 

In the static case where the central banks keeps MB constant, the credit expansion of banks reduces liquidity, since as mentioned above, whenever a loan is made and come back as deposit on the banking system, a certain fraction need to be retained as required reserve, though the total amount of deposit, and hence M2, increases.

In the same token, as the purchase of securities increases, again in the static case where the central bank keeps MB constant, M2 increase while liquidity decreases; see the description of the repo debacle in 2019 in [money market linear](../fixed-income/instruments/gsac/money-market-linear.ipynb). ## What are the conventional monetary policy tools used by the Fed?

### Open Market Operations

The Fed conducts most of its open market operations in Treasury securities, because the market for these securities is the most liquid and has the largest trading volume. The Fed does this by interacting with primary dealers over TRAPS (Trading Room Automated Processing System).

- Dynamic open market operations: intended to change the level of reserves and monetary base
- Defensive open market operations: intended to offset movements in other factors that affect reserve and monetary base. 
    - To offset transient issues (such as check floats or deposit of Treasury; in the case of PBoC, there is also offsetting changes to funds outstanding for foreign exchange), the Fed mainly use repo and RRP.
    - Sometimes outright transactions are conducted, over TRAPS

The Fed is considered to have full control over OMO as for its timing and quantity. It is also flexible and precise: the Fed can reverse course at any time. It can also be implemented more quickly than e.g. reserve requirement, since banks will have to given advanced notice so that they can adjust the required reserve calculation.

### Discount Loan Facilities

- Discount Window
    - Primary credit: mainly as a tool to conduct monetary policy: it lends to healthy banks for very short maturities (usually overnight). The interest rate is usually set higher than the Fed fund target, usually by 100 basis point - during the Crisis, it was once dropped to 25bps to encourage bank obtain this liquidity. By an arbitrage argument, this discount rate is an upper bound of the effective Fed fund rate (see [money market linear](../fixed-income/instruments/gsac/money-market-linear.ipynb))
    - Secondary credit is given to banks that are in financial trouble and are experiencing severe liquidity problems. The interest rate on secondary credit is set at 50 basis points above the discount rate - supposed to be penalizing.
    
- Lender of last resort

Note that the Fed's lender-of-last-resort is still needed in presence of the FDIC insurance.
- First, it is important to recognize that the FDICs insurance fund amounts to about 1% of the total amount of deposits held by banks. If a large number of bank failures occurred simultaneously, the FDIC would not be able to cover all the depositors losses.

- A loss of confidence in the banking system could still lead to runs on banks from the largedenomination depositors, and bank panics could still occur despite the existence of the FDIC.

During the recent Financial Crisis, the Fed not only act as lender of last resort to the banks, but also the broader financial system as a whole.

### Reserve Requirements

This method was once a powerful way of affecting the money supply and interest rates but is rarely used today (probably because it is too blunt.)

### Interest on Reserves (IOER)

The Fed has usually set IOER below the Fed fund target. The Fed also has used it as a monetary policy tool coming out of the Crisis.
- Back in 2015, when the Fed starts to hike rate, it did so by increasing IOER and using its floor effect on Fed fund, not by engaging large scale OMO to drain reserves.
- In the liquidity drain in the hiking cycle before 2020, the Fed delibrately hiked IOER 5bps less than the Fed fund target, to discourage banks holding excess reserve.

## What are in China's interest rate toolbox?



Since China's financial market is not yet fully liberated, between the tools to control the interest rate and the quantity of money (as shown above), the PBoC still has a lot of tools to control the latter.



#### Different Functions of the Tools

A recap of the existing liquidity tools mentioned in previous note: reverse repo, MLF, targeted medium-term lending facility (TMLF), standing lending facility (SLF) and treasury cash deposit auction.

- Among the tools, MLF and TMLF have the longest tenors. In our view, the PBOC regards them as the means to inject longer-term liquidity to aid lending growth, when the timing or market situation are not appropriate for RRR cuts.
- The others are mostly to manage short-term liquidity volatility, 7-day reverse repo being the main go-to tool.
- SLF works as a supplement to the reverse repo; it is used to help individual banks that encounter liquidity shortage, but cannot get adequate funding in the money market.
- Unlike the above-mentioned tools, the operation of treasury cash deposit auction depends on the liquidity condition of the Ministry of Finance, and is not fully controlled by the PBOC. Another difference is that the injection through these auctions does not increase the central banks balance sheet, meaning it does not increase base money.

#### Timing of Operation

- The PBOC gauges cash demand of eligible counterparties -- mainly major banks -- every working day, and offers funds via reverse repo. But the offered volume depends on PBOCs judgement, and may be different from the banks requests. The central bank tends to inject large volume of funds at some specific periods: these include, but are not limited to, 1-2 weeks prior to New Year and Spring Festival holidays, end of each quarter, and the tax payment seasons (mainly mid-April, May and July).
- MLF is mostly offered when previous ones mature, but the PBOC may present a larger size than the amount set to mature, so as to incrementally increase liquidity injection.
- TMLF was formally launched this year. So far the PBOC has kept offers limited to once a quarter - three batches of TMLFs have been offered in January, April and July.
- SLF is operated at PBOCs discretion when banks need urgent liquidity support.
- Treasury cash deposit auctions are held based on the cash level of the Ministry of Finance. Usually it can be performed up to 2 times a month.

#### Liquidity Tools vs. RRR Adjustment

Both a RRR cut and an injection with the above-mentioned tools can increase the level of overall liquidity in the banking sector. That said, some differences exist:

- A RRR adjustment does not change the size of the PBOCs or commercial banks balance sheets, as opposed to the reverse repo, MLF, TMLF and SLF.
- Cash released by RRR cuts does not require any collateral from commercial banks, but the tools above do. For example, high-quality bonds or loans are required as collateral when banks get cash via MLF from the PBOC.
- There is no maturity to cash released by RRR cuts. When the PBOC wants to tighten market conditions, it will raise the RRR. But cash injected through liquidity operations are subjected to a maturing date.
- Funds released by RRR cuts are viewed as of no cost to the banks, while those obtained via liquidity tools are subject to certain interest rates. For example, 7-day reverse repo costs 2.55% and one-year MLF costs 3.3%.
- RRR cuts have stronger implication to the market on the PBOCs policy stance than liquidity tools.
Despite the differences, RRR cuts and liquidity tools are often used together to balance impact. For example, the PBOC may withdraw short-term cash from the market following a cut in required reserve ratio.

In recent years, as the endeavor to liberate the financial system carries on, the PBoC also manage a lieu of interest rates. These interest rates can also be considered as the policy instrument of PBoC.
- In the open market, theres the 7-day reverse repo (007) and the Medium-term Lending Facility (MLF), two pivotal policy rates set by the central bank for its loans to commercial banks. 
- For lending to the real economy, theres the new Loan Prime Rate (LPR), introduced last year in part to help lower borrowing costs of companies and households.
- In the context of LIBOR replacement, the PBoC wants to elevate a key rate in the interbank market, the Depository-Institutions Repo Rate (DR) to benchmark status. Its used in the interbank market, where financial institutions sell yuan between themselves, and is based on actual trades. Started in 2014, it is the daily, weighted average of pledged repo transactions, where a borrower pledges bonds in exchange for a loan. There are 11 different DR tenors ranging from overnight to one-year, so compared to some RFRs overseas which are only overnight, it may be better suited to underpinning a wider range of financial products. At the moment the most actively traded are the overnight and the 7-day rates.
