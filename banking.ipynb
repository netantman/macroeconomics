### Liability, or funding source

- Checkable deposits
    - Checkable deposits are bank accounts that allow the owner of the account to write checks to third parties. Checkable deposits include all accounts on which checks can be drawn, and thus is payable on demand. 
    - They are usually the lowest-cost source of bank funds, because depositors are willing to forgo some interest in exchange for access to a liquid asset that they can use to make purchases immediately.
    - With the advent of new, more attractive financial instruments, such as money market deposit accounts, the share of checkable deposits in total bank liabilities has shrunk over time.
    - Note that the cost incurred to serve these accounts not only include interest payments, but also the logistic and maintenance costs such as processing, preparing statement, paying tellers, etc.

- Nontransaction deposits
    - Includes negotiable CDs (NCDs), time deposits and savings account.
    - Nowadays the more important funding source for banks.
    - NCDs are typically bought by corporations or other banks. Largedenomination CDs are negotiable; like bonds, they can be resold in a secondary market before they mature. For this reason, negotiable CDs are held by corporations, money market mutual funds, and other financial institutions as alternative assets to Treasury bills and other short-term bonds.

- Borrowings
    - Interbank market for Fed fund reserves.
    - Discount loans from the Fed directly

- Bank Capital, or equity

### Asset, or funding use

- Reserves, or vault cash
    - required reserve
    - excess reserve

- Cash Items in Process of Collection

- Deposits at Other Banks: in China, banks take advantage of this item to circumvent regulatory restriction on loans.

- Securities
    - Treasuries and munis: note that banks are not allowed to hold stocks.
    - Short-term U.S. government securities are called secondary reserves.

- Loans
    - Largest categories of loans for commercial banks are commercial and industrial loans made to businesses and real estate loans. 

- Other Assets
    - Physical capital: bank buildings, computers, and other equipments.

## What justifies the existence of banks?

Basic banking is to engage in **asset transformation**,  by selling liabilities with one set of characteristics (a particular combination of liquidity, risk, size, and return) and using the proceeds to buy assets with a different set of characteristics.

Besides the argument that banks take liquidity risk in engaging asset-liability management, the other argument justifying the existence of banks is that they reduce information asymmetry: simply put, by vetting credit-worthy borrower, banks are compensated for this generated information (and since loans are private, it does not lead to other free-riding, and thus the business model holds).

## How do banks management interest rate risks?

- **Gap analysis**: the amount of rate-sensitive liabilities is subtracted from the amount of rate-sensitive assets
    - **Maturity bucket approach**: to measure the gap for several maturity buckets and use the expected life for mortgages perhaps.
    - **Standardized gap analysis**: account for the differing degrees of rate sensitivities among rate-sensitive assets and liabilities.

- **Duration analysis**: match the duration of assets and liabilities, derivatives, such as swaps can help banks manage interest rate risk efficiently

## What off-balance-sheet activity banks engage in?

### Loan Sales

A loan sale, also called a secondary loan participation, involves a contract that sells all or part of the cash stream from a specific loan and thereby removes the loan so that it is no longer an asset on the banks balance sheet.

### Generation of Fee Income

Banks receive fees for providing specialized services to their customers. 
- making foreign exchange trades on a customers behalf, 
- servicing a mortgage-backed security
- collecting interest and principal payments and then paying them out
- guaranteeing debt securities such as bankers acceptances (by which the bank promises to make interest
and principal payments if the party issuing the security cannot)
- providing backup lines of credit
    - loan commitment
    - credit lines with overdraft previlege: essentially, write themselves a loan
    - standby letter of credit
    - credit lines for underwriting securities

## What are loan loss provisions?

Without getting into too much of accounting, it is the same as loan loss allowance. It reflects the estimate of banks on the expected loss of their loans, and hold some fund in each period for this expected loss. On banks' balance sheet it manifests as a reduction of asset, though the loss has not actually occured. Loan loss provision can accumulate and becomes some sort of reserve. To ensure safety on banking, there are regulation on the ratio of loan loss reserve balance over total loan amount: in China recently this ratio cannot be lower than 2.5%. Banks usually put more in the reserve in good times and less in bad, as a way to smooth out the economic cycle.
