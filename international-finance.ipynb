## How is foreign exchange traded?

The foreign exchange market is organized as an over-the-counter market in which several hundred dealers (mostly banks) stand ready to buy and sell deposits denominated in foreign currencies. Because these dealers are in constant telephone and computer contact, the market is very competitive; in effect, it functions no differently from a centralized market.

An important point to note is that although banks, companies, and governments talk about buying and selling currencies in foreign exchange markets, they do not take a fistful of dollar bills and sell them for British pound notes. Rather, most trades involve the buying and selling of bank deposits denominated in different currencies. So when we say that a bank is buying dollars in the foreign exchange market, what we actually mean is that the bank is buying deposits denominated in dollars. The volume in this market is colossal, exceeding $5 trillion per day.

Trades in the foreign exchange market consist of transactions in excess of $1 million for the wholesale market, and this is usually the rate that is quoted. For retail transactions, the rate is less favorable.

## What is PPP?

Theory of Purchasing Power Parity. It states that exchange rates between any two currencies will adjust to reflect changes in the price levels of the two countries. The theory of PPP is simply an application of the law of one price to national price levels rather than to individual prices.

The PPP can also be understood in terms of real exchange rate, which is the rate at which domestic goods can be exchanged for foreign goods. The real exchange rate indicates whether a currency is relatively cheap or not (as is the application of the Economists' big mac index). PPP predicts that the real exchange rate is always equal to 1, so that the purchasing power of the dollar is the same as that of other currencies.

Though PPP presents a compeling equilibrium-type argument in the long run, empirically it has poor predictive power over the short run. This is because the assumptions of PPP are often violated in the short term: (a) goods are identical in different places; (b) transportation costs are zero and (c) there is no trade barriers (actually many goods are not traded across borders at all).

## How is exchange rate managed, and how does it affect the domestic monetary policy?

Central banks manage exchange rate by using international reserve. An **unsterilized intervention** in which domestic currency is bought/sold (from public) and foreign assets are sold/bought (for reserve) leads to a fall/rise in international reserves, a fall/rise in the money supply, and an appreciation/depreciation of the domestic currency. 

A **sterilized intervention**, by not altering the quantity of domestic money supply and the interest rate difference, has almost no effect on the exchange rate, even though it does changes the amount of foreign securities relative to domestic securities in the hands of the public, called a portfolio balance effect.  Through this effect, the central bank might be able to influence the interest differential between domestic and foreign assets, which in turn affects the relative expected return on domestic assets. Empirical evidence has not revealed this portfolio balance effect to be significant. However, a sterilized intervention could indicate what central banks want to happen to the future exchange rate and so might provide a signal about the course of future monetary policy.

The impact of foreign exchange intervention on monetary policy is best summarized in the **Mundell-Fleming triangle**, also called the **impossible trinity**: monetary authorities can only have two out of the three things below.
